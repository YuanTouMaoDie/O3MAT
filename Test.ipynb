```python
import pandas as pd
import numpy as np

# 定义需要的列名
usecols = ['ROW', 'COL', 'Timestamp', 'vna_ozone', 'avna_ozone', 'evna_ozone', 'model']

# 定义开始和结束时间，时间跨度为2011年1月1日至1月7日
start_date = pd.Timestamp('2011-01-01 00:00:00')
end_date = pd.Timestamp('2011-01-02 23:00:00')
date_rng = pd.date_range(start=start_date, end=end_date, freq='H')

# 生成ROW和COL的组合
rows, cols = np.meshgrid(np.arange(1, 300), np.arange(1, 460))
rows = rows.flatten()
cols = cols.flatten()

# 每个网格对应一个星期的24小时数据，需要复制日期范围
total_data_points = len(rows) * len(date_rng)
all_rows = np.repeat(rows, len(date_rng))
all_cols = np.repeat(cols, len(date_rng))
all_timestamps = np.tile(date_rng, len(rows))

# 生成模拟的臭氧数据和模型数据，这里简单使用随机数
vna_ozone = np.random.uniform(0, 100, size=total_data_points)
avna_ozone = np.random.uniform(0, 100, size=total_data_points)
evna_ozone = np.random.uniform(0, 100, size=total_data_points)
model = np.random.uniform(0, 100, size=total_data_points)

# 组合数据为字典
data_dict = {
    'ROW': all_rows,
    'COL': all_cols,
    'Timestamp': all_timestamps,
    'vna_ozone': vna_ozone,
    'avna_ozone': avna_ozone,
    'evna_ozone': evna_ozone,
    'model': model
}

# 创建DataFrame
simulated_df = pd.DataFrame(data_dict)

# 将数据保存为CSV文件，以便后续测试使用
simulated_df.to_csv('/DeepLearning/mnt/shixiansheng/data_fusion/output/Test/simulated_data.csv', index=False)

print("模拟数据已生成并保存为 simulated_data.csv")
```
